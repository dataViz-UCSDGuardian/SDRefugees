```python
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
plotly.tools.set_credentials_file(username='ayatamin', api_key='d9z1mogc01')
```

```python
f_2015 = pd.read_excel('RefugeesInAm2015.xlsx')
f_2015
```

```python
country = []
ca_country = []
total = []
for i in range(1,125):
    if( f_2015.ix[4,:][i] > 200):# or f_2015.ix[51,:][i] > 100 ):
        ca_country.append(f_2015.ix[4,:][i])
        total.append(f_2015.ix[51,:][i])
        country.append(f_2015.columns.values[i])
```

```python
ca = go.Bar( x = country, y = ca_country, name = 'California')
usa = go.Bar( x = country, y = total, name = 'USA')
data = [ca, usa]
layout = go.Layout(barmode='stack')
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='refugees in cali')
```
