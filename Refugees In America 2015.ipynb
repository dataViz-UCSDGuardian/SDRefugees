```python
import pandas as pd
import numpy as np
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
plotly.tools.set_credentials_file(username='ayatamin', api_key='d9z1mogc01')
```

```python
f_2015 = pd.read_excel('RefugeesInAm2015NEW.xlsx')
f_2015
```

```python
country = []
ca_country = []
total = []
percentages = []
ones = []
CA_row = 4;
USA_row = 49;
for i in range(1,80):
    if( f_2015.ix[CA_row,:][i] > 100):# or f_2015.ix[51,:][i] > 100 ):
        c = f_2015.ix[CA_row,:][i]
        t = f_2015.ix[USA_row,:][i]
        ct = f_2015.columns.values[i]
        
        ca_country.append(c)
        total.append(t-c)
        country.append(ct)
        ones.append(1 - (c/t))
        percentages.append( c/t )
```

```python
import colorlover as cl
color = cl.scales['9']['seq']['YlGnBu']
```

```python
fig = {'data': [{'labels': country,
            'values': ca_country,
            'type': 'pie',
            'marker': {'colors': 
                       [color[3], 
                        color[0], 
                        color[4], 
                        color[7], 
                        color[8], 
                        color[6], 
                        color[2], 
                        color[5], 
                        color[1] ]}}],
    'layout': {'title': 'Where Refugees in California are from'}
     }

py.iplot(fig,  filename='where cali refugees are from for 2015')
```

```python
ca = go.Bar( x = country, y = percentages, name = 'California', marker = dict( color= color[5]))
usa = go.Bar( x = country, y = ones, name = 'USA', marker = dict( color= color[2]))
data = [ca, usa]
layout = go.Layout(barmode='stack',  
                   title = 'What Proportion of Refugees in America live in California?')
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='percent refugees in cali out of total')
```
